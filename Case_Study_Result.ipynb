```python
import pandas as pd
from sklearn.linear_model import LinearRegression as LR
```

```python
df = pd.read_csv("Life Expectancy.csv")
```

```python
df = df.dropna()
nndata = df[df["Country"] == "Afghanistan"]
nndata
```

# **Task 1**

BMI is selected because it can affect the life expectancy.

Source: [Obesity in adulthood and its consequences for life expectancy: a life-table analysis](https://pubmed.ncbi.nlm.nih.gov/12513041/)

```python
y1 = nndata["Life expectancy "]
x1 = nndata[[" BMI "]]

model1 = LR()
model1.fit(x1, y1)

print(f"R^2 = {model1.score(x1, y1)}")
print("Adjusted R^2 = "+str(1-15*(1-model1.score(x1, y1))/14))
print(f"Intercept = {model1.intercept_}")
print(f"Coefficients = {model1.coef_}")
```

# **Task 2**

BMI and GDP are selected because they can affect the life expectancy.

Source: [Socioeconomic development and life expectancy relationship: evidence from the EU accession candidate countries](https://genus.springeropen.com/articles/10.1186/s41118-019-0071-0)

```python
y2 = nndata["Life expectancy "]
x2 = nndata[[" BMI ", "GDP"]]

model2 = LR()
model2.fit(x2, y2)

print(f"R^2 = {model2.score(x2, y2)}")
print("Adjusted R^2 = "+str(1-15*(1-model2.score(x2, y2))/13))
print(f"Intercept = {model2.intercept_}")
print(f"Coefficients = {model2.coef_}")
```

# **Task 2**

BMI, GDP, Population are selected because they can affect the life expectancy.

Source: [Socioeconomic development and life expectancy relationship: evidence from the EU accession candidate countries](https://genus.springeropen.com/articles/10.1186/s41118-019-0071-0)

```python
y3 = nndata["Life expectancy "]
x3 = nndata[[" BMI ", "GDP", "Population"]]

model3 = LR()
model3.fit(x3, y3)

print(f"R^2 = {model3.score(x3, y3)}")
print("Adjusted R^2 = "+str(1-15*(1-model3.score(x3, y3))/12))
print(f"Intercept = {model3.intercept_}")
print(f"Coefficients = {model3.coef_}")
```
